```python
import pandas as pd
import json

# --- Method 1: Basic Load ---
# This keeps 'conversations' as a list and 'metadata' as a dictionary in their columns
df = pd.read_json("raw_dataset.jsonl", lines=True)
print("Basic columns:", df.columns.tolist())

# --- Method 2: Flattened Load (Recommended) ---
# This expands the 'metadata' dictionary into distinct columns (e.g., 'metadata.domain')
with open("raw_dataset.jsonl", "r", encoding="utf-8") as f:
    data = [json.loads(line) for line in f]

df_flat = pd.json_normalize(data)
print("\nFlattened columns:", df_flat.columns.tolist())

# Preview the data
print(df_flat.head())
```
